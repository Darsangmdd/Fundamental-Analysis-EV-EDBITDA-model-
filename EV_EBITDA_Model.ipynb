```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('https://www.moneycontrol.com/financials/relianceindustries/consolidated-ratiosVI/RI#RI').text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
table = soup.find('table' , class_="mctable1")
```

```python
tabs = []
for tr in table.find_all('tr')[1:]:
  tabs.append({
            'KEY RATIOS (in Rs. Cr.)': tr.find_all('td')[0].text,
            'MAR 21': tr.find_all('td')[1].text,
            'MAR 20': tr.find_all('td')[2].text,
            'MAR 19': tr.find_all('td')[3].text,
            'MAR 18': tr.find_all('td')[4].text,
            'MAR 17': tr.find_all('td')[5].text,
                        })
print(tabs)
```

```python
import pandas as pd
df = pd.DataFrame(tabs)
df.set_index('KEY RATIOS (in Rs. Cr.)', inplace = True)
df
```

```python
df.dtypes
```

```python
df2 = df.iloc[[35,37], :]
df2.columns
```

```python
df2['MAR 21'] = df2['MAR 21'].str.replace(',','').astype('float')
df2['MAR 20'] = df2['MAR 20'].str.replace(',','').astype('float')
df2['MAR 19'] = df2['MAR 19'].str.replace(',','').astype('float')
df2['MAR 18'] = df2['MAR 18'].str.replace(',','').astype('float')
df2['MAR 17'] = df2['MAR 17'].str.replace(',','').astype('float')
df2
```

```python
 EBITDA = df2.iloc[0,:].values / df2.iloc[1,:].values
 EBITDA = EBITDA.tolist()
 print(EBITDA)
 df2.loc['EBITDA'] = EBITDA
 df2
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
df2
```

```python
Growth_in_EBITDA = []
for i in range(len(EBITDA)-1):
  value = ((EBITDA[i]-EBITDA[i+1])/EBITDA[i+1])*100
  Growth_in_EBITDA.append(value)
  i+=1 
print(Growth_in_EBITDA)
import numpy as np
print('Mean growth in EBITDA in the past three years = ',np.mean(Growth_in_EBITDA[0:3]))
```

```python
print('Expected EBITDA in coming year = {} crores'.format((EBITDA[0]*(100+np.mean(Growth_in_EBITDA[0:3])))/100))
```

```python
source2 = requests.get('https://www.moneycontrol.com/india/stockpricequote/refineries/relianceindustries/RI').text
```

```python
soup2 = BeautifulSoup(source2, 'lxml')
table = soup2.find('table' , class_="peer_tbl")

tabs = []
for tr in table.find_all('tr')[1:]:
  tabs.append({
            'Company Name': tr.find_all('td')[0].text,
            'LTP': tr.find_all('td')[1].text,
            'MCap': tr.find_all('td')[3].text,
                        })
print(tabs)
```

```python
df5 = pd.DataFrame(tabs)
df5.set_index('Company Name', inplace = True)
```

```python
df5['LTP'] = df5['LTP'].str.replace(',','').astype('float')
df5['MCap'] = df5['MCap'].str.replace(',','').astype('float')
df5
```

```python
Outstanding_shares = df5.iloc[0,]/df5.iloc[0,0]
```

```python
print('The number of Outstanding shares is {} crores'.format(Outstanding_shares))
```

```python
Target_Price = (Forecasted_EV)/ Outstanding_shares
Entry_Price = 0.75*Target_Price
print('The Target price according to the model is Rs.{}'.format(round(Target_Price,2)))
print('The Entry price according to the model is Rs.{}'.format(round(Entry_Price,2)))
```
